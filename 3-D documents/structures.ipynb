# Plotting Compounds Using 3-D Structures Data


```python
import pandas as pd                           #Importing Libraries
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

Reading the csv file and taking a look at the first 5 compounds data.

```python
df= pd.read_csv('database.csv', encoding='cp1252')
print (df.head())
```

Taking df1 as Name column and df2 as STRUCTURE column.

```python
df1=df['Name']
print (df1.head())
```

```python
df2=df['STRUCTURE']
print (df2.head())
```

When a particular compound's structure data is viewed, it is noticed that there are three numbers unique after V2000. These three numbers are unique for every compound,so they are plotted on a 3-D plot along with the compond name. 

```python
print (df2[0])
```

Two arrays are taken one to store the compound name and the other one to store the three numbers.There are compounds which don't have their structures yet and also have their structure data in a slightly different order. Those compounds are neglected for now. The suitable structures data are splitted and inserted in the array.

```python
array1=[]
array2=[]
print ("Compounds which are not in required format or NaN")
for j in range(len(df2)):
    try:
        tf=df2[j]
        if tf[0]=='C':  #Neglection of different format structure data.
            print (j)
        else:
            t=tf[46:72]  #Splitting the three Numbers
            #print(t)
            t=t.split()
            array1.append(t)
            array2.append(df1[j])
    except:
        print(j)
```

The 3-D plot is plotted using the three numbers from array1 and their corresponding label is the compound name from array2. The compounds which cause error while plotting are printed out.

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
print('error while plotting compound:')
for k in range(len(array2)):
    try:
        ax.text(float(array1[k][0]),float(array1[k][1]),float(array1[k][2]), str(array2[k]), color='red')
    except:
        print (k)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```
